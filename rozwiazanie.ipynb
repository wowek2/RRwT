## **1. Tło  historyczne:** <a class="anchor" id="tlo-historyczne"></a>


Kotlina Jelenigórska wraz z przylegającymi do niej Karkonoszami oraz Górami Izerskimi, ze względu na spore pokłady kwarcu, przez wieki była zagłębiem hutnictwa szkła. Huty te nazywano *'wędrownymi'*, bo gdy w jednym miejscia zaczęło brakować drewna do opalania pieców, to przenoszono je dalej - a właścwie to wyżej, coraz głębiej w dziewicze, górskie lasy. Pod koniec *XVI* wieku skutki rabunkowej gospodarki leśnej, doprowadziły do całkowitego wylesienia południowych zboczy Karkonoszy. Usunięcie ekosystemu, który wcześniej działał jak gąbka, która absorbowała wodę, a później oddawała ją powoli przez miesiące, doprowadziło do regularnych powodzi nawiedzających Kotlinę Jeleniogórską oraz tereny leżące przy rzece Bóbr.

W nocy z $29$ na $30$ lipca $1897$ roku doszło do ogromnej ulewy. Jak odnotowano na stacji Nová Louka w ciąu jednego dnia spadło aż $345$ milmietrów deszczu (średnia roczna suma opadów w polsce to $600$ mm). Na wylesionych zboczach potoki z sekundy na sekundę pęczniały. O godzinie dwudziestej pierwszej stany wody przekroczyły próg alarmowy, a potem nadeszła fala. Wylały wszystkie rzeki. Poziom Bobru podniósł się o ponad $5$ metrów, a Kotlnia Jeleniogórska zamieniłą się w ogromny akwen. We wleniu Bóbr zalał rynek i okalające go kamienice. W wyniku powodzi zginęło ok. $120$ osób. Straty materianle po stronie niemieckiej oszaczowano na $10$-$18$ milionów marek.

Tragedia z $1897$ roku uruchomiła machinę jakiej Śląsk jeszcze nie widział. Katastrofa nie była lokalnym problemem, ale sprawą całego Cesarstwa Niemieckiego. Nadprezydent prowincji Hermann von Hatzfeldt zwrócił się do wybitnego inżyniera Otto Intze o przygotowanie systemu zbiorników retencyjnych mających chronić przed kolejnymi powodziami. Profesor przedstawił cesarzowi Willhelmowi II osobiście plan, a już w $1900$ roku parlament pruski uchwalił *Śląską Ustawę o ochronie przeciwpowodziowej* - dokument przewidywał regulację siedmiu głównych rzek: Osobłogi, Nysy Kłodzkiej, Bystrzycy Kaczawy, Bobru, Kwisy oraz Nysy Łużyckiej. Całkowity koszt programu oszaczowano na $39$ milionów marek - sumę astronomiczną jak na tamte czasy.

Miejsce na główną zaporę kaskady wybrano w Pilchowicach. Tam gdzie Bóbr tworzy łagodne zakole między Górą Zamkową ($350$ m.n.p.m), a Górą Korzec ($438$ m.n.p.m). W szczytowym momencie na placu budowy pracowało siedmiusetpięćdziesięciu robotników. Dyscyplina była żelazna: za spóźnienie się, pijaństwo czy palenie tytoniu w trakcie pracy groziły dotkliwe kary finansowe. $16$ listopada $1912$ roku nastąpiło uroczyste oddanie, w tamtym momencie największej zapory Europy. W uroczystościach uczestniczył sam cesarz - Wilhelm II. Liczby robiły wrażenie: wysokość sześćdziesiąt dwa metry od fundamentu, długość korony dwieście siedemdziesiąt metrów. Powstałe Jezioro Pilchowickie zajęło dwieście czterdzieści hektarów i pomieściło pięćdziesiąt milionów metrów sześciennych wody. Całkowity koszt budowy zapory wyniósł koło ośmiu milionów marek.

W ramach projektu pomiędzy Jelenią Górą, a Pilchowicami wzniesiono także: kamienną zaporę jazową Bobrowice I ($1924$-$1925$) oraz zaporę Wrzeszczyn ($1926$-$1927$), będącą jazem dwuprzęsłowym.

W $1945$ roku obiekty przeszły pod polską administację. Bóbr jednak nie dał o sobie zapomnieć. W lipcu $1997$ roku - niemal dokładnie w stulecie powodzi z $1897$ roku - natura chciała sprawdzić czy dzieło Intze'go przetrwa próbę czasu. Woda przelała się przez koronę, ale zapora wytrzymała. We wrześniu $2024$ roku przepływ Bobru przekroczył sześćset metrów sześciennych na sekundę - więcej niż w $1997$ roku. A zapora w Pilchowicach znów stanęła na wysokości zadania (choć podobnie jak w $1997$ roku woda przelała się przez koronę). W styczniu $2026$ roku rozpoczął się gruntowny remont konstrukcji. Po raz pierwszy od lat siedemdziesiątych zbiornik ma być całkowicie opróżniony. Prace potrwają do końca $2027$ roku i kosztować będą ponad dziewięćdziesiąt trzy miliony złotych ubu

### **2.1 Opis i topologia kaskady**
Analizowany system hydrotechniczny położony jest na rzecze Bóbr. Rzeka opuszczając **Kotlinę Jeleniogórską**, wpływa w wąwóz **Borowego Jaru**. System składa się z trzech głównych stopni wodnych połączonych w układ szeregowy. Oznacza to że odpływ ze stopnia górnego stanowi dopływ do stopnia dolnego.

Jak można zauważyć mamy trzy zbiorniki retencyjne. Stopień wodny **Borowice I** jest obiektem o małej pojemności i charakteryzuje się szybką reakcją na zmiany dopływu branego bezpośrednio z wodowskazu w Jeleniej Górze. Drugim obiektem jest **Zapora Wrzeszczyn** o średniej pojemności. Dopływem tego zbiornika jest zrzut z zapory **Borowice I** oraz nieznaczny **potok Wądół**. Ostatnim obiektem jest **Zapora Pilchowice** - główny zbiornik systemu charakteryzujący się bardzo dużą pojemnością. Jego dopływ stanowi zrzut z **Zapory Wrzesczyn** oraz znaczący dopływ boczny **rzeki Kamienica**.

<span style="color:red">**UWAGA:**</span> W naszym modelu pomijamy stopnie wodne **Borowice IV** oraz **Borowice II**. Stopnie te nie posiadają zdolności retencyjnych, a woda przepływa przez nie swobodnie.

### **2.2 Sformułowanie problemu badawczego**

**Celem pracy** jest skonstruowanie modelu matematycznego, który na drodze symulacji numerycznej odtworzy nieznane przebiegi wewnętrzne i pozwoli zminimalizować błąd pomiędzy wymodelowanym, a rzeczywistym wypływie w Pilchowicach.

Dysponujemy danymi pomiarowymi jedynie na krańcach tego systemu:
* **Stan wejściowy $Q_{in}(t)$:** Przepływ na wodowskazie w Jeleniej Górze.
* **Stan wyjściowy $Q_{out}(t):$** Dane z wodowskazu znajdującego się zaraz za zaporą w Pilchowicach.

Nie posiadamy bezpośrednich pomiarów stanów wewnętrznych, takich jak aktualne objętości wody w poszczególnych zbiornikach, czy natężenie przepływów bocznych (Kamienica, Wądół).

### **2.3 Założenia uproszczające model**
W celu sformułwania naszego modelu, zastosujemy szereg założeń upraszczających rzeczywistą naturę zjawisk hydrotechnicznych.
* Traktujemy każdy zbiornik, jako punkt materialny o zmiennej masie. Pomijamy wszelkie modelowanie ruchów turbulentnych i laminarnych, koncentrując się jedynie na globalnym bilansie masy zbiornika.
* **Nieściśliwość cieczy:** Zakładamy stałą gęstość wody $\rho = \text{const}$, co pozwala na sformułowanie bilansu w kategoriach objętościowych($m^3$), a nie masowych ($kg$).
* **Pomijamy niektóre zjawiska:** Dla modelu działającego, w krótkim oknie czasowym możemy pominąć uwzględnienie zjawisk tj. parowanie, inflirtacja wody w grunt, itd. 
* **Modelowanie dopływów bocznych:** Ze względu na brak pomiarów zlewni pośrednich wpadających do systemu, przyjmujemy hipotezę wyjaśnioną w dalszej części.
* **Aproksymacja odcinków rzecznych:** Odcinki koryta rzeki pomiędzy zaporami nie są traktowane jako kanały o opóźneinu transportowym $\tau$ (prowadziłoby to do **DDE**),   zamiast tego stosujemy model **retencji liniowej** wyjaśniony w dalszej części.

## 3. **Modelowanie matematyczne**

### 3.1 **Przestrzeń stanu i wektor stanu**
**Stanem systemu** hydrotechnicznego w chwili $t$ jest zbiór wielkości fizycznych, które jednoznacznie określają ilość wody zgromadzoną w poszczególnych elementach naszego układu. <br> Definiujemy wektor stanu $\mathbf{x} \in \mathbb{R}^5$:

$$\mathbf{x(t)} =
\begin{bmatrix}
x_{1} \\
x_{2} \\
x_{3} \\
x_{4} \\
x_{5} 
\end{bmatrix} =
\begin{bmatrix}
V_M(t) \\
V_{R_{1}}(t) \\
V_W(t) \\
V_{R_{2}}(t) \\
V_P(t) 
\end{bmatrix}
\text{, gdzie } 
\begin{array}{l}
V_M \text{ -- objętość jeziora Modre} \\
V_{R_{1}} \text{ -- retencja koryta na odcinku łączącym jez. Modre oraz jez. Wrzeszczyńskie} \\
V_W \text{ -- objętoś jeziora Wrzeszczyńskiego} \\
V_{R_{2}} \text{ -- retencja koryta na odcinku łączącym jez. Wrzeszczyńskie oraz jez. Pilchowickie} \\
V_P \text{ -- objętość jeziora Pilchowickiego}
\end{array}$$

Przestrzeń stanów dopuszczalnych $\mathcal{X}$ jest ograniczonym podzbiorem $\mathbb{R}^5$. Ograniczenia te wynikają z fizyki oraz konstrukcji obiektów - nieujemna objętość oraz maskymalna pojemność zbiorników.

$$\mathcal{X} = \{ \mathbf{x} \in \mathbb{R}^5 : 0 \le x_i \le V_{max_{i}}, \quad \forall i \in \{1, 2, 3, 4, 5\} \}$$
<span style="color:red">Uwaga</span>: *Dla odcinków rzecznych $x_2, x_4$ teoretyczna pojemność odpowiada pojemności retencyjnej doliny zalewowej i jest znacznie wyższa niż pojemnność koryta.*

Przyjęcie definicji stanu jako retencji, a nie przepływu jest kluczowe, ponieważ pozwala na bezpośrednie zastosowanie zasady zachowania masy.

### 3.2 **Wprowadzenie równań**
Dynamika każdego elementu kaskady (zarówno zbiornika, jak i odcinka rzeki) wynika z fundamentalnego prawa zachowania masy, które w hydrologii przyjmuje postać **równania ciągłości**.

$$\frac{dV}{dt} = Q_{in}(t) - Q_{out}(t)$$

, gdzie $Q_{in}$ to całkowity dopływ, a $Q_{out}$ to całkowity odpływ w chwili $t$.<br> O ile dla zbiorników retencyjnych równanie to stosujemy bezpośrednio, o tyle dla odcinków rzecznych konieczne jest zdefiniowanie związku między retencją a przepływem.
