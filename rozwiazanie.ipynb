## **1. Tło historyczne i hydrotechniczne Kotliny Jeleniogórskiej** <a class="anchor" id="tlo-historyczne"></a>


### **Hutnictwo szkła i wylesienie Sudetów**

Kotlina Jeleniogórska wraz z przylegającymi do niej Karkonoszami oraz Górami Izerskimi, ze względu na znaczne pokłady kwarcu, przez wieki była ważnym ośrodkiem hutnictwa szkła. Funkcjonowały tu tzw. huty „wędrowne”, które przenoszono w głąb górskich lasów w miarę wyczerpywania lokalnych zasobów drewna wykorzystywanego do opalania pieców.

Pod koniec XVI wieku skutki rabunkowej gospodarki leśnej doprowadziły do niemal całkowitego wylesienia południowych zboczy Karkonoszy. Usunięcie naturalnego ekosystemu, który wcześniej działał jak gąbka – magazynując wodę i oddając ją stopniowo – spowodowało gwałtowne zwiększenie odpływu powierzchniowego oraz regularne powodzie w Kotlinie Jeleniogórskiej i w dolinie rzeki Bóbr.


### **Powódź z lipca 1897 roku**

W nocy z 29 na 30 lipca 1897 roku doszło do jednej z największych katastrof hydrologicznych w historii regionu. Na stacji meteorologicznej Nová Louka zanotowano dobowy opad w wysokości 345 mm, co stanowiło ponad połowę średniej rocznej sumy opadów w Polsce.

Na wylesionych zboczach Karkonoszy górskie potoki w krótkim czasie zamieniły się w rwące strumienie. Około godziny 21:00 stany wód przekroczyły progi alarmowe, po czym nastąpiła gwałtowna fala powodziowa. Wylały wszystkie rzeki, a poziom Bobru podniósł się miejscami o ponad 5 metrów.

Kotlina Jeleniogórska zamieniła się w rozległy akwen, a w Lwówku Śląskim rzeka zalała rynek i otaczające go kamienice. W wyniku powodzi zginęło około 120 osób, natomiast straty materialne po stronie niemieckiej oszacowano na 10–18 milionów marek.

### **System ochrony przeciwpowodziowej Otto Intzego**

Katastrofa z 1897 roku stała się impulsem do podjęcia bezprecedensowych działań w skali całego Cesarstwa Niemieckiego. Nadprezydent prowincji śląskiej Hermann von Hatzfeldt zlecił wybitnemu inżynierowi Otto Intzemu opracowanie systemu ochrony przeciwpowodziowej opartego na kaskadzie zbiorników retencyjnych.

Projekt został osobiście przedstawiony cesarzowi Wilhelmowi II, a w 1900 roku parlament pruski uchwalił tzw. Śląską Ustawę o ochronie przeciwpowodziowej. Dokument ten przewidywał regulację oraz zabezpieczenie siedmiu głównych rzek regionu: Osobłogi, Nysy Kłodzkiej, Bystrzycy, Kaczawy, Bobru, Kwisy oraz Nysy Łużyckiej.

Całkowity koszt programu oszacowano na około 39 milionów marek, co stanowiło sumę ogromną jak na ówczesne realia.

### **Budowa zapory w Pilchowicach**

Miejsce budowy głównej zapory kaskady Bobru wybrano w Pilchowicach, w naturalnym zwężeniu doliny pomiędzy Górą Zamkową (350 m n.p.m.) a Górą Korzec (438 m n.p.m.).

W szczytowym okresie prac na placu budowy zatrudnionych było około 750 robotników, a dyscyplina pracy była niezwykle surowa – za spóźnienia, pijaństwo czy palenie tytoniu w trakcie robót groziły wysokie kary finansowe.

Uroczyste oddanie zapory do użytku nastąpiło 16 listopada 1912 roku w obecności cesarza Wilhelma II. W momencie ukończenia była to największa zapora w Europie:

wysokość: 62 m (od fundamentu),

długość korony: ok. 270 m.

Utworzone Jezioro Pilchowickie zajęło powierzchnię około 240 ha i mogło zgromadzić blisko 50 mln m³ wody. Całkowity koszt budowy zapory wyniósł około 8 milionów marek
(Pilchowice Dam).

Dalsza rozbudowa kaskady Bobru

W ramach realizacji programu ochrony przeciwpowodziowej pomiędzy Jelenią Górą a Pilchowicami powstały również inne obiekty hydrotechniczne:

Bobrowice I – kamienna zapora jazowa (1924–1925),

Wrzeszczyn – zapora pełniąca funkcję jazu dwuprzęsłowego (1926–1927).

Zapora w XX i XXI wieku

Po 1945 roku wszystkie obiekty hydrotechniczne w dolinie Bobru przeszły pod polską administrację. Rzeka wielokrotnie przypominała jednak o swoim potencjale niszczycielskim.

W lipcu 1997 roku, podczas tzw. powodzi tysiąclecia, woda przelała się przez koronę zapory w Pilchowicach, jednak konstrukcja wytrzymała ekstremalne obciążenie.

We wrześniu 2024 roku przepływ Bobru przekroczył 600 m³/s, czyli więcej niż podczas wezbrania w 1997 roku. Również tym razem zapora spełniła swoją funkcję ochronną, mimo przelewu wody przez koronę.

### **Współczesna modernizacja**

W styczniu 2026 roku rozpoczął się kompleksowy remont zapory w Pilchowicach. Po raz pierwszy od lat siedemdziesiątych XX wieku zbiornik ma zostać całkowicie opróżniony.

Prace modernizacyjne zaplanowano do końca 2027 roku, a ich koszt szacowany jest na ponad 93 miliony złotych. Remont ma na celu dostosowanie ponadstuletniej konstrukcji do współczesnych standardów bezpieczeństwa hydrotechnicznego oraz dalszą ochronę doliny Bobru przed ekstremalnymi wezbraniami.

## **3. Modelowanie Matematyczne** <a class="anchor" id="modelowanie"></a>

Celem niniejszego rozdziału jest sformułowanie modelu dynamicznego kaskady zbiorników wodnych na rzece Bóbr. System ten jest układem naczyń połączonych, gdzie stan każdego kolejnego zbiornika zależy od historii przepływów w zbiornikach poprzedzających oraz charakterystyki zlewni pośrednich.

### **3.1. Bilans masy (Równanie ciągłości)**

Podstawą modelowania każdego zbiornika retencyjnego jest prawo zachowania masy, które w hydrologii przyjmuje postać równania ciągłości przepływu. [cite_start]Dla dowolnego zbiornika $i$, zmiana objętości wody $V_i$ w czasie $t$ jest różnicą pomiędzy sumą dopływów a sumą odpływów[cite: 1, 31]:

$$\frac{dV_i}{dt} = Q_{in, i}(t) - Q_{out, i}(h_i(t)) + Q_{add, i}(t)$$

gdzie:
* $V_i$ – objętość wody w zbiorniku [$m^3$],
* $Q_{in, i}$ – dopływ główny z rzeki (lub zrzut z poprzedniego zbiornika) [$m^3/s$],
* $Q_{out, i}$ – całkowity odpływ ze zbiornika, będący funkcją poziomu wody $h_i$ [$m^3/s$],
* $Q_{add, i}$ – dopływy boczne ze zlewni bezpośredniej oraz dopływów niesterowalnych (np. potoki Wądół, Kamienica).

### **3.2. Geometria zbiorników (Relacja $h-V$)**

Ponieważ w równaniach różniczkowych operujemy zmianą objętości ($dV/dt$), a odpływ fizyczny zależy od poziomu spiętrzenia ($h$), konieczne jest zdefiniowanie relacji geometrycznej wiążącej te wielkości.
Zakładając uproszczony model doliny rzecznej, kształt zbiornika aproksymujemy jako bryłę o zmiennym przekroju (np. odwrócony ostrosłup ścięty lub klin). [cite_start]W praktyce inżynierskiej często stosuje się przybliżenie, w którym powierzchnia lustra wody $A(h)$ zmienia się liniowo lub potęgowo wraz z wysokością[cite: 31].

Przyjmijmy model potęgowy relacji wysokości do objętości:
$$h(V) = H_{max} \cdot \left( \frac{V}{V_{max}} \right)^{\frac{1}{\beta}}$$
Gdzie parametr $\beta$ zależy od kształtu doliny (dla idealnego stożka/ostrosłupa $\beta=3$, dla prostopadłościanu $\beta=1$, dla doliny rynnowej $\beta \approx 2$). W naszym modelu parametr ten zostanie dobrany eksperymentalnie na podstawie danych technicznych zapór (powierzchnia zalewu przy $H_{max}$).

### **3.3. Charakterystyka odpływu ($Q_{out}$)**

Odpływ z zapory jest sumą przepływu przez turbiny elektrowni wodnej oraz przepływu przez przelewy awaryjne (koronę zapory), gdy poziom wody przekroczy stan maksymalny.

1.  **Przepływ przez turbiny ($Q_{turb}$):**
    Traktujemy go jako sterowalny do momentu osiągnięcia przepływności maksymalnej turbin. W modelu uproszczonym, przy braku harmonogramu pracy elektrowni, zakładamy pracę ciągłą przy dostępnym poziomie wody:
    $$Q_{turb}(h) = \min(Q_{turb\_max}, \alpha \sqrt{h})$$
    [cite_start]Wzór ten bazuje na prawie Torricellego, gdzie prędkość wypływu $v = \sqrt{2gh}$[cite: 1].

2.  **Przepływ przez przelew powierzchniowy ($Q_{spill}$):**
    Gdy poziom wody przekroczy rzędną korony przelewu ($h > h_{max}$), woda przelewa się swobodnie. [cite_start]Zjawisko to opisuje wzór na wydatek przelewu ostrego (formuła Francisa/Poleniego)[cite: 31]:
    $$Q_{spill}(h) = C_d \cdot L \cdot (h - h_{max})^{\frac{3}{2}}$$
    gdzie:
    * $C_d$ – współczynnik wydatku (zależny od kształtu korony),
    * $L$ – długość korony przelewu,
    * $h_{max}$ – wysokość progu przelewu.

Ostatecznie:
$$Q_{out}(h) = Q_{turb}(h) + \mathbb{I}(h > h_{max}) \cdot Q_{spill}(h)$$

### **3.4. Topologia i nieznane dopływy**

Modelowany system składa się z trzech węzłów. Problemem badawczym jest brak danych pomiarowych dla dopływów bocznych (rzeka Kamienica, potok Wądół). Aby rozwiązać ten problem, zastosujemy metodę analogii hydrologicznej, skalując znany dopływ $Q_{JeleniaGora}$ współczynnikiem $\lambda$, proporcjonalnym do powierzchni zlewni lub średniego opadu.

**Układ równań modelu:**

1.  **Zbiornik Bobrowice I (Jezioro Modre):**
    $$\frac{dV_{Modre}}{dt} = Q_{in}(t) - Q_{out, Modre}(h_{Modre})$$

2.  **Zbiornik Wrzeszczyn:**
    $$\frac{dV_{Wrz}}{dt} = Q_{out, Modre}(t - \tau_1) + \lambda_{Wadol} \cdot Q_{in}(t) - Q_{out, Wrz}(h_{Wrz})$$
    *Uwaga: $\tau_1$ reprezentuje opóźnienie transportowe na odcinku rzecznym.*

3.  **Zbiornik Pilchowice:**
    $$\frac{dV_{Pil}}{dt} = Q_{out, Wrz}(t - \tau_2) + \lambda_{Kamienica} \cdot Q_{in}(t) - Q_{out, Pil}(h_{Pil})$$

Celem symulacji jest minimalizacja błędu pomiędzy $Q_{out, Pil}$ a rzeczywistym pomiarem w profilu Pilchowice, poprzez optymalizację parametrów $\lambda$.
